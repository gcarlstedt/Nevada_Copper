```python
#Security
#Concat multiple excel sheets at once from directory file

import numpy as np
import pandas as pd
import glob, os
import datetime

pd.options.mode.chained_assignment = None

#Path to folder with files we want
path = r'C:\Users\gcarlstedt\Documents\Work - Nevada Copper\Security Logs\Python_Version'
#Get all the files from path that .xlsx
all_files = glob.iglob(os.path.join(path, "*.xlsx"))
#For loop to read files in all_files and concat to a dataframe
df = pd.concat((pd.read_excel(file) for file in all_files), ignore_index=True, sort=False, verify_integrity=True)

#Make a copy of original DataFrame
df1 = df.copy()
#Select certain columns
data = df1[['Date','Person Name','Reader Name']]

#Checking which columns have null values
data.isnull().any()
#They all do
#Figure out which rows
data[data['Reader Name'].isnull()]
#Keep only the rows with at least 3 non-NA values
data.dropna(thresh=3,inplace=True)
#Check again for null values
data.isnull().any()
#No more null values

#Convert 'Date' column to datetime 
data['Date']= data['Date'].astype('datetime64[ns]') 

#Clean up 'Person Name' column --> Don't want ID#
#Keep all letters until position -6
data['Person Name'] = data['Person Name'].map(lambda x: str(x)[:-7])

data
```

```python
#Replace Cemlist copy with update list
copy['Name'] = update

#Filter 'Person Name' column in data with those in the new copy['Name'] column 
condensedlist = data.loc[data['Person Name'].isin(copy['Name'])]

#Reset index of this new condensed list with cem employees
condensedlist.reset_index(inplace=True, drop=True)

#Sort Columns
condensedlist.sort_values(by=['Date'],ascending=True,inplace=True)

#To account for only night shifts, break apart into 3 separate dataframes to filter and then
#recombine

#Night Shift for May 31st to June 1st

#Made a mistake, thought May only had 30 days instead of 31 --> Not good
#Shift May 30th to May 31st
may = condensedlist[(condensedlist['Date'] < '2019-05-30 23:59:59')]
may['Date'] = may.Date + datetime.timedelta(days=1)
maynight = may[(may['Date'] >'2019-05-31 12:00:00') & (may['Reader Name'] == 'Entry Reader')]

#Night Shift for July 31st to August 1st
august = condensedlist[(condensedlist['Date'] > '2019-07-31 23:59:59')]
augustnight = august[(august['Date'] < '2019-08-01 12:00:00') & (august['Reader Name'] == 'Exit Reader')]

# nunique() counts all unique values in column etc.
#Get main body of dates excluding may and august dates
Body = condensedlist[(condensedlist['Date'] > '2019-05-30 23:59:59') & (condensedlist['Date'] < '2019-07-31 23:59:59')]

#Recombine all dataframes with filtered Night shifts
final = maynight.append(Body).append(augustnight)
#Sort by name and then date
final.sort_values(by=['Person Name','Date'], ascending=True,inplace=True)
#Reindex 
final.reset_index(drop=True,inplace=True)

#Check for null values in columns
final.isnull().any()
#None --> Proceed with calculating work hours

final
```

```python
test = final.copy()

#Transpose 'Reader Name' into two columns labeled Entry and Exit?
#Attempt with pivot table or groupby
#Use pivot for non numeric data
testpivot = test.pivot(columns = 'Reader Name', values='Date')

testpivot.insert(0, 'Person Name', test['Person Name'])
#testpivot['shifted'] = testpivot.groupby(['Person Name'])['Exit Reader'].shift(-1)
testpivot['shifted'] = testpivot.groupby(['Person Name'])['Exit Reader'].transform(lambda x:x.shift(-1))
testpivot['Hours Worked'] = testpivot['shifted'] - testpivot['Entry Reader']
#testpivot.dropna(thresh=3,inplace=True)

#new = testpivot[testpivot['Person Name'] == 'Higby, Arthur']
new = testpivot[['Person Name', 'Entry Reader', 'shifted','Hours Worked']].copy()
new.dropna(thresh=2,inplace=True)

new.rename(columns={'shifted': 'Exit Reader'}, inplace=True)
new.reset_index(drop=True,inplace=True)

print('Total Hours Worked for June & July:', testpivot['Hours Worked'].sum())
print('Unaccounted Shifts (Estimate):',len(new) - new['Hours Worked'].count())
print('Unaccounted Hours (Assuming ~12 Hr shifts):', 58*12)
#new.style.hide_index()
new[new['Hours Worked'].isnull()].style.hide_index()
```
